```python
products_clean = products.dropna(subset=['product_category_name_english'])
print("Продукти преди почистване:", len(products))
print("Продукти след почистване:", len(products_clean))
```

```python
# Категориите в products, които ги има като текст, но нямат превод
products_with_category = products[products['product_category_name'].notna()]
missing_translation = products_with_category[products_with_category['product_category_name_english'].isna()]

print("Уникални категории без превод:")
print(missing_translation['product_category_name'].unique())
print("\nБрой продукти засегнати:", len(missing_translation))
```

```python
# Взимаме уникален списък от продукти с категория (не искаме дубликати от order_products)
unique_products = products_clean[['product_id', 'product_category_name_english']].drop_duplicates()

print("Уникални продукти:", len(unique_products))
unique_products.head()
```

```python
category_encoded = pd.get_dummies(unique_products['product_category_name_english'])

print("Форма на матрицата:", category_encoded.shape)
category_encoded.head()
```

```python
import torch

# Превръщаме True/False матрицата в PyTorch tensor от числа (0.0 / 1.0)
category_matrix = torch.tensor(category_encoded.values.astype(float), dtype=torch.float32)

print("Форма на tensor-а:", category_matrix.shape)
print("Пример - първите 5 стойности на продукт 0:", category_matrix[0][:5])
```

```python
import torch.nn.functional as F

# Смятаме similarity между ВСИЧКИ двойки продукти наведнъж
similarity_matrix = F.cosine_similarity(
    category_matrix.unsqueeze(1),   # (32328, 1, 71)
    category_matrix.unsqueeze(0),   # (1, 32328, 71)
    dim=2
)

print("Форма на similarity матрицата:", similarity_matrix.shape)
```

```python
def get_similar_products(product_index, top_n=5):
    # Векторът само на избрания продукт
    target_vector = category_matrix[product_index].unsqueeze(0)  # (1, 71)
    
    # Similarity между този един продукт и ВСИЧКИ останали
    similarities = F.cosine_similarity(target_vector, category_matrix, dim=1)  # (32328,)
    
    # Топ N най-близки (без самия продукт)
    top_indices = similarities.argsort(descending=True)[1:top_n+1]
    
    return top_indices, similarities[top_indices]

# Тестваме с продукт с индекс 2 (спомни си - това беше 'sports_leisure')
indices, scores = get_similar_products(2, top_n=5)
print("Similarity стойности:", scores)
print("\nПрепоръчани продукти:")
print(unique_products.iloc[indices.numpy()])
```

```python
def get_similar_products(product_index, top_n=5):
    target_vector = category_matrix[product_index].unsqueeze(0)
    similarities = F.cosine_similarity(target_vector, category_matrix, dim=1)
    
    # Сортираме всички индекси по similarity
    sorted_indices = similarities.argsort(descending=True)
    
    # Изрично премахваме самия продукт по index, не по позиция
    sorted_indices = sorted_indices[sorted_indices != product_index]
    
    top_indices = sorted_indices[:top_n]
    return top_indices, similarities[top_indices]

# Тест пак
indices, scores = get_similar_products(2, top_n=5)
print("Similarity стойности:", scores)
print("\nПрепоръчани продукти:")
print(unique_products.iloc[indices.numpy()])
```

```python
from transformers import pipeline

sentiment_model = pipeline(
    "sentiment-analysis",
    model="nlptown/bert-base-multilingual-uncased-sentiment"
)
```

```python
# Тестваме с прост пример на португалски
result = sentiment_model("Recebi bem antes do prazo estipulado.")
print(result)
```

```python
# По-силно позитивен и по-силно негативен пример за сравнение
result_positive = sentiment_model("Produto excelente, superou minhas expectativas! Recomendo muito.")
result_negative = sentiment_model("Produto péssimo, veio quebrado e o vendedor não respondeu.")

print("Позитивен пример:", result_positive)
print("Негативен пример:", result_negative)
```

```python
# Разбиваме sentiment_result (dict) на две отделни колони - по-лесно за работа
reviews_with_text['predicted_label'] = reviews_with_text['sentiment_result'].apply(lambda x: x['label'])
reviews_with_text['predicted_confidence'] = reviews_with_text['sentiment_result'].apply(lambda x: x['score'])

# Запазваме резултата на диска
reviews_with_text.to_csv('../data/reviews_with_sentiment.csv', index=False)

print("Запазено успешно!")
print(reviews_with_text[['review_score', 'predicted_label', 'predicted_confidence']].head(10))
```

```python
# Свързваме sentiment резултатите с продукти/категории (през order_items)
sentiment_with_category = order_products.merge(
    reviews_with_text[['order_id', 'predicted_label', 'predicted_confidence']], 
    on='order_id', how='inner'
)

print("Редове:", len(sentiment_with_category))
sentiment_with_category[['product_category_name_english', 'predicted_label', 'predicted_confidence']].head()
```

```python
# Извличаме числото от label ("5 stars" -> 5)
sentiment_with_category['predicted_score'] = sentiment_with_category['predicted_label'].str[0].astype(int)

category_sentiment = sentiment_with_category.groupby('product_category_name_english')['predicted_score'].agg(['mean', 'count'])
category_sentiment = category_sentiment[category_sentiment['count'] >= 30].sort_values('mean')

print("Категории с най-негативен sentiment (текст):")
print(category_sentiment.head(10))
```
